```python
img = cv2.imread('/content/sample_data/1.jpg')
print(img.shape)
plt.imshow(img)
```

```python
RGBimg = img[...,::-1]
RGBimg2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(121);plt.imshow(RGBimg)
plt.subplot(122);plt.imshow(RGBimg2)
```

```python
Channel_1,Channel_2,Channel_3 = cv2.split(RGBimg)
plt.imshow(Channel_1)
```

```python
img_rs = cv2.resize(RGBimg,(128,128))
plt.subplot(121);plt.imshow(RGBimg);plt.title('Original Image')
plt.subplot(122);plt.imshow(img_rs);plt.title('Resized Image')
```

```python
height,width = RGBimg.shape[:2]
print(height,width)
scale = 0.5
img_rs = cv2.resize(RGBimg,(int(width*scale),int(height*scale)))
plt.subplot(121);plt.imshow(RGBimg);plt.title('Original Image')
plt.subplot(122);plt.imshow(img_rs);plt.title('Resized Image')
```

```python
crp_img = RGBimg[:100,:100]
plt.subplot(121);plt.imshow(RGBimg);plt.title('Original Image')
plt.subplot(122);plt.imshow(crp_img);plt.title('Cropped Image')
```

```python
flip_img = cv2.flip(RGBimg,1)
plt.subplot(121);plt.imshow(RGBimg);plt.title('Original Image')
plt.subplot(122);plt.imshow(flip_img);plt.title('Flipped Image')
```

```python
matrix = np.ones(RGBimg.shape,dtype=np.uint8)*50
bright_img = cv2.add(RGBimg,matrix)
dark_img = cv2.subtract(RGBimg,matrix)
plt.subplot(131);plt.imshow(RGBimg);plt.title('Original Image')
plt.subplot(132);plt.imshow(bright_img);plt.title('Brightened Image')
plt.subplot(133);plt.imshow(dark_img);plt.title('Darkened Image')
```

```python
matrix = np.ones(RGBimg.shape,dtype=np.uint8)
high_contrast_img = cv2.multiply(RGBimg,matrix*1.2,dtype=cv2.CV_8UC3)
low_contrast_img = cv2.multiply(RGBimg,matrix*0.8,dtype=cv2.CV_8UC3)
plt.subplot(131);plt.imshow(RGBimg);plt.title('Original Image')
plt.subplot(132);plt.imshow(high_contrast_img);plt.title('High Contrast Image')
plt.subplot(133);plt.imshow(low_contrast_img);plt.title('Low Contrast Image')
```

```python
img_bgr = cv2.cvtColor(RGBimg, cv2.COLOR_RGB2BGR)
new_img = cv2.line(img_bgr, (100, 100), (200, 200), (0, 0, 255), 2)
#First param must be BGR formatted
new_img = cv2.rectangle(new_img, (100, 100), (200, 200), (0, 255, 0), 2)
new_img = cv2.circle(new_img, (100, 100), 50, (255, 0, 0), 2)
plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)) # Convert back to RGB for displaying with matplotlib
```

# Machine learning
- Machine learning is the study of algorithms that can learn from experience
- As a machine learning algorithm accumulates more experience, typically in the form of observational data or interactions with an environment, its performance improves.
- In the currently dominant approach to machine learning, we do not attempt to design a system explicitly to recognize wake words. Instead, we define a flexible program whose behavior is determined by a number of parameters.
Then we use the dataset to determine the best possible parameter values, i.e., those that improve the performance of our program with respect to a chosen performance measure.

- In machine learning, the learning is the process by which we discover the right setting of the knobs for coercing the desired behavior from our model. In other words, we train our model with data.
- This sort of problem, where we try to predict a designated unknown label based on known inputs given a dataset consisting of examples for which the labels are known, is called supervised learning.
- no matter what kind of machine learning problem we tackle:


1.   The data that we can learn from.
2.   A model of how to transform the data.
3.   An objective function that quantifies how well (or badly) the model is doing.
4.   An algorithm to adjust the models parameters to optimize the objective function.



## Data
Each example (or data point, data instance, sample) typically consists of a set of attributes called features (sometimes called covariates or inputs), based on which the model must make its predictions. In supervised learning problems, our goal is to predict the value of a special attribute, called the label (or target), that is not part of the models input.
